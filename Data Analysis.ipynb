```python
def group_name(userid):
    control_group = {'d', 'f', '1', '2', '3', '4', '5', '8'}  # Using a set for faster lookup
    test_group = {'a', 'b', 'c', 'e', '0', '6', '7', '9'}     # Also strings for consistency
    
    if not isinstance(userid, str):
        userid = str(userid)  # Ensure it's a string
    
    if not userid:  # Handle empty input
        return "unknown"  # or raise an error
    
    last_char = userid[-1].lower()  # Case-insensitive check
    
    if last_char in control_group:
        return "control"
    elif last_char in test_group:
        return "test"
    else:
        return "unknown"  # Handle unexpected characters
```

```python
campaign_data['group'] = campaign_data['userID'].apply(group_name)
```

```python
campaign_data.head()
```

```python
campaign_data.info()
```

#### Count of userid in each group
